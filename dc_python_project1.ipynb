```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Loading the csv file
netflix_df = pd.read_csv("C://Users//Admin//Documents//sahi_studies//netflix_data.csv")
```

```python
netflix_df.shape
```

```python
#filtering the dataset to remove TV shows
netflix_subset = netflix_df.drop(netflix_df[netflix_df['type'] == "TV Show"].index)
```

```python
netflix_subset.shape
```

```python
#subsetting the data with specific columns
netflix_movies = netflix_subset[["title","country","genre","release_year","duration"]]
```

```python
#movies having duration shorter than 60 mins
short_movies = netflix_movies[netflix_movies['duration'] <60]
```

```python
short_movs
```

```python
#Assigning different colors to various genre in the netflix movies list
colors = []
for lable, row in netflix_movies.iterrows():
    if row['genre'] == "Children":
        colors.append('yellow')
    elif row['genre'] == "Documentaries":
       colors.append('blue')
    elif row['genre'] == "Stand-Up":
        colors.append('green')
    else:
        colors.append('red')
print(colors)
```

```python
#data visualization
fig = plt.figure(figsize=(12,8))
plt.scatter(netflix_movies['release_year'], netflix_movies['duration'], c=colors)
plt.xlabel("Release Year")
plt.ylabel("Duration (min)")
plt.title("Movie Duration by Year of Release")
plt.show()
```
